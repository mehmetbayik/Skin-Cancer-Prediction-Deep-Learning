```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gzip
```

```python
# Plots for different batch sizes
model1 = '13-12-15-19'
model2 = '13-12-15-22'
model3 = '13-12-15-24'
history1 = pd.read_csv(f'model-comparison/{model1}/history.csv')
history2 = pd.read_csv(f'model-comparison/{model2}/history.csv')
history3 = pd.read_csv(f'model-comparison/{model3}/history.csv')

# Plot Validation Accuracies in one plot
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(6, 2)
index = np.arange(1,history1.shape[1])
ax.plot(index, history1.iloc[3][1:], label='Batch Size: 1')
ax.plot(index, history2.iloc[3][1:], label='Batch Size: 64')
ax.plot(index, history3.iloc[3][1:], label='Batch Size: 50000')
ax.set_title('Validation Accuracy History')
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.legend()
plt.savefig('batch-size-comparison.png')
plt.show()
```

```python
# Plots for different weight initializations
model1 = '13-12-15-15'
model2 = '13-12-15-16'
model3 = '13-12-15-17'
history1 = pd.read_csv(f'model-comparison/{model1}/history.csv')
history2 = pd.read_csv(f'model-comparison/{model2}/history.csv')
history3 = pd.read_csv(f'model-comparison/{model3}/history.csv')

# Plot Validation Accuracies in one plot
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(6, 2)
index = np.arange(1,history1.shape[1])
ax.plot(index, history1.iloc[3][1:], label='Weight Init: Zero')
ax.plot(index, history2.iloc[3][1:], label='Weight Init: Uniform')
ax.plot(index, history3.iloc[3][1:], label='Weight Init: Normal')
ax.set_title('Validation Accuracy History')
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.legend()
plt.savefig('weight-init-comparison.png')
plt.show()
```

```python
# Plots for different learning rates
model1 = '13-12-15-25'
model2 = '13-12-15-26'
model3 = '13-12-15-27'
model4 = '13-12-15-28'
history1 = pd.read_csv(f'model-comparison/{model1}/history.csv')
history2 = pd.read_csv(f'model-comparison/{model2}/history.csv')
history3 = pd.read_csv(f'model-comparison/{model3}/history.csv')
history4 = pd.read_csv(f'model-comparison/{model4}/history.csv')

# Plot Validation Accuracies in one plot
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(6, 2)
index = np.arange(1,history1.shape[1])
ax.plot(index, history1.iloc[3][1:], label='LR: 0.1')
ax.plot(index, history2.iloc[3][1:], label='LR: 0.001')
ax.plot(index, history3.iloc[3][1:], label='LR: 0.0001')
ax.plot(index, history4.iloc[3][1:], label=' LR: 0.00001')
ax.set_title('Validation Accuracy History')
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.legend()
plt.savefig('learning-rate-comparison.png')
plt.show()
```
